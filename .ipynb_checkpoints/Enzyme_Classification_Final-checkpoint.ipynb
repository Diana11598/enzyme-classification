# Enzyme Classification using Machine Learning

This notebook builds a binary classifier that distinguishes enzyme sequences from
non-enzyme sequences using classical protein sequence features
(physicochemical properties, amino-acid composition, dipeptide composition)
and tree-based ML models.

**Pipeline overview**
1. Load and balance the dataset
2. Extract sequence-based features (AAC, DPC, physicochemical)
3. Compare baseline models
4. Compare feature sets
5. Feature selection
6. Hyperparameter tuning of the final model
7. Evaluation (classification report, ROC-AUC, confusion matrix)
8. Model interpretability (feature importance, SHAP)



````python
import random
import numpy as np

random.seed(42)
np.random.seed(42)

import os

os.makedirs("figures", exist_ok=True)
````

## Part 1: Data Loading, Balancing & Feature Extraction

````python
# ==========================================================
# Enzyme Classification using Machine Learning
# Part 1: Data Loading & Feature Extraction
# ==========================================================

# ==========================
# Import Libraries
# ==========================

import pandas as pd
import numpy as np

from Bio import SeqIO
from Bio.SeqUtils.ProtParam import ProteinAnalysis


# ==========================
# Load FASTA Files
# ==========================

data = []

# Positive class (Enzymes)
for record in SeqIO.parse("enzymes.fasta", "fasta"):
    data.append({
        "id": record.id,
        "seq": str(record.seq),
        "label": 1
    })

# Negative class (Non-Enzymes)
for record in SeqIO.parse("non_enzymes.fasta", "fasta"):
    data.append({
        "id": record.id,
        "seq": str(record.seq),
        "label": 0
    })

df = pd.DataFrame(data)

print("Original Dataset")
print(df["label"].value_counts())

# ==========================
# Balance Dataset
# ==========================

enzyme = df[df["label"] == 1]

non_enzyme = (
    df[df["label"] == 0]
    .sample(n=len(enzyme), random_state=42)
)

df = pd.concat(
    [enzyme, non_enzyme],
    ignore_index=True
)

print("\nBalanced Dataset")
print(df["label"].value_counts())

# ==========================
# Sequence Cleaning
# ==========================

amino_acids = "ACDEFGHIKLMNPQRSTVWY"

valid_amino = set(amino_acids)

dipeptides = [
    a + b
    for a in amino_acids
    for b in amino_acids
]

df["seq"] = df["seq"].apply(
    lambda seq: "".join(
        aa for aa in seq
        if aa in valid_amino
    )
)
````

**Note:** undersampling the majority class here throws away a lot of real
non-enzyme sequences (161,219 → 49,490). That's a valid, simple way to handle
imbalance, but it costs you data. Two alternatives worth knowing about for a
future iteration: keep the full dataset and use `class_weight="balanced"` in the
models, or oversample the minority class (e.g. SMOTE on the numeric features,
*after* the train/test split) instead of undersampling.

````python
# ==========================
# Feature Extraction
# ==========================

feature_list = []

labels = []

for i, sequence in enumerate(df["seq"]):

    if len(sequence) < 5:
        continue

    protein = ProteinAnalysis(sequence)

    features = {}

    # -----------------------------
    # Physicochemical Features
    # -----------------------------

    features["aromaticity"] = protein.aromaticity()

    features["instability_index"] = protein.instability_index()

    features["isoelectric_point"] = protein.isoelectric_point()

    helix, turn, sheet = protein.secondary_structure_fraction()

    features["sec_struct_helix"] = helix

    features["sec_struct_turn"] = turn

    features["sec_struct_sheet"] = sheet
    # -----------------------------
    # Amino Acid Composition (AAC)
    # -----------------------------

    for aa in amino_acids:

        features[aa] = sequence.count(aa) / len(sequence)

    # -----------------------------
    # Dipeptide Composition (DPC)
    # -----------------------------

    dp_count = {
        dp: 0
        for dp in dipeptides
    }

    for j in range(len(sequence) - 1):

        pair = sequence[j:j+2]

        if pair in dp_count:
            dp_count[pair] += 1

    for dp in dipeptides:

        features[dp] = (
            dp_count[dp] /
            (len(sequence) - 1)
        )

    feature_list.append(features)

    labels.append(df.iloc[i]["label"])
````

**Performance note:** this loop is pure Python over ~99k sequences with a
nested O(len × 400) dipeptide count each — it's the slowest cell in the
notebook. Not a bug, but if you re-run this a lot, computing the dipeptide
counts with `collections.Counter(zip(sequence, sequence[1:]))` instead of
scanning the 400-item `dipeptides` list for every position would cut this
down substantially.

````python
# ==========================
# Create Feature DataFrame
# ==========================

feature_df = pd.DataFrame(feature_list)

feature_df["label"] = labels

print(feature_df.shape)

feature_df.head()
````

````python
# ==========================
# Prepare Machine Learning Dataset
# ==========================

X = feature_df.drop(
    columns="label"
)

y = feature_df["label"]

print("Number of Features:", X.shape[1])

print("Number of Samples:", X.shape[0])
````

````python
# ==========================
# Train-Test Split
# ==========================

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print(X_train.shape)

print(X_test.shape)
````

## Part 2: Baseline Model Comparison

````python
# ==========================================================
# Part 2: Baseline Model Comparison
# ==========================================================

import time

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import (
    StratifiedKFold,
    cross_val_score
)

from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import (
    RandomForestClassifier,
    GradientBoostingClassifier,
    ExtraTreesClassifier,
    HistGradientBoostingClassifier
)

from sklearn.neighbors import KNeighborsClassifier

# ==========================
# Define Baseline Models
# ==========================

pipelines = {

    "Logistic Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("model", LogisticRegression(max_iter=2000))
    ]),

    "KNN": Pipeline([
        ("scaler", StandardScaler()),
        ("model", KNeighborsClassifier())
    ]),

    "Gradient Boosting": Pipeline([
        ("model", GradientBoostingClassifier(
            random_state=42
        ))
    ]),

    "Hist Gradient Boosting": Pipeline([
        ("model", HistGradientBoostingClassifier(
            random_state=42
        ))
    ]),

    "Random Forest": Pipeline([
        ("model", RandomForestClassifier(
            n_estimators=100,
            random_state=42,
            n_jobs=-1
        ))
    ]),

    "Extra Trees": Pipeline([
        ("model", ExtraTreesClassifier(
            n_estimators=100,
            random_state=42,
            n_jobs=-1
        ))
    ])

}


# ==========================
# Cross Validation
# ==========================

cv = StratifiedKFold(
    n_splits=3,
    shuffle=True,
    random_state=42
)

results = {}
# ==========================
# Evaluate Models
# ==========================

for name, model in pipelines.items():

    print("=" * 50)

    print(f"Running {name}")

    start = time.time()

    scores = cross_val_score(
        model,
        X_train,
        y_train,
        cv=cv,
        scoring="roc_auc",
        n_jobs=-1
    )

    end = time.time()

    print(f"Mean ROC-AUC : {scores.mean():.4f}")

    print(f"Std ROC-AUC  : {scores.std():.4f}")

    print(f"Time         : {(end-start)/60:.2f} minutes")

    results[name] = scores.mean()
````

````python
# ==========================
# Model Comparison Table
# ==========================

baseline_results = pd.DataFrame({
    "Model": results.keys(),
    "ROC_AUC": results.values()
})

baseline_results = baseline_results.sort_values(
    by="ROC_AUC",
    ascending=False
)

baseline_results.reset_index(
    drop=True,
    inplace=True
)

baseline_results

# ==========================
# Plot Baseline Results
# ==========================

import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))

plt.bar(
    baseline_results["Model"],
    baseline_results["ROC_AUC"]
)

plt.xticks(rotation=20)

plt.ylabel("ROC-AUC")

plt.title("Baseline Model Comparison")

plt.grid(alpha=0.3)


plt.savefig("figures/baseline_models.png", dpi=300, bbox_inches="tight")

plt.show()
````

````python
print(
    "Best Baseline Model:",
    baseline_results.iloc[0]["Model"]
)
````

## Feature Set Comparison (AAC vs AAC+DPC vs AAC+DPC+Physicochemical)

````python
# ==========================================================
# Feature Set Comparison
# ==========================================================

phys_columns = [
    "aromaticity",
    "instability_index",
    "isoelectric_point",
    "sec_struct_helix",
    "sec_struct_turn",
    "sec_struct_sheet"
]

aac_columns = list(amino_acids)

dpc_columns = dipeptides

X_aac = feature_df[aac_columns]

X_aac_dpc = feature_df[
    aac_columns + dpc_columns
]

X_all = feature_df[
    aac_columns +
    dpc_columns +
    phys_columns
]

# ==========================
# Evaluate Feature Set
# ==========================

def evaluate_feature_set(X, y):
    
    X_train_fs, X_test_fs, y_train_fs, y_test_fs = train_test_split(
        X,
        y,
        test_size=0.20,
        stratify=y,
        random_state=42
    )

    model = RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        n_jobs=1
    )

    scores = cross_val_score(
        model,
        X_train_fs,
        y_train_fs,
        cv=3,
        scoring="roc_auc",
        n_jobs=1
    )

    return scores.mean()
````

````python
# run comparison
auc_aac = evaluate_feature_set(
    X_aac,
    y
)

auc_aac_dpc = evaluate_feature_set(
    X_aac_dpc,
    y
)

auc_all = evaluate_feature_set(
    X_all,
    y
)
# DataFrame for result
feature_results = pd.DataFrame({
    "Feature Set":[
        "AAC",
        "AAC + DPC",
        "AAC + DPC + Phys"
    ],
    "Number of Features":[
        20,
        420,
        426
    ],
    "ROC_AUC":[
        auc_aac,
        auc_aac_dpc,
        auc_all
    ]
})

feature_results
````

````python
plt.figure(figsize=(7,5))

plt.bar(
    feature_results["Feature Set"],
    feature_results["ROC_AUC"]
)

plt.ylabel("ROC-AUC")

plt.title("Comparison of Biological Feature Sets")

plt.ylim(0.80,1.0)

plt.grid(alpha=0.3)

# FIX: filename had no extension and contained spaces
plt.savefig("figures/feature_set_comparison.png", dpi=300, bbox_inches="tight")
plt.show()
````

## Part 3: Feature Selection

````python
# ==========================================================
# Part 3: Feature Selection
# ==========================================================

from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

# ==========================
# Train Feature Selector
# ==========================

selector_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

selector_model.fit(X_train.values, y_train)

# ==========================
# Select Important Features
# ==========================

selector = SelectFromModel(
    estimator=selector_model,
    threshold="mean",
    prefit=True
)

X_train_selected = selector.transform(X_train)

X_test_selected = selector.transform(X_test)

selected_features = X_train.columns[
    selector.get_support()
]

print("Original Features :", X_train.shape[1])

print("Selected Features :", X_train_selected.shape[1])
````

**Important methodological note:** `selector_model` is fit once on the whole
of `X_train`, and the resulting `selected_features` are then reused inside the
3-fold cross-validation below. Strictly speaking, this is a mild form of
leakage — the feature-selection step "saw" every training fold before CV even
started, so the reported CV score is very slightly optimistic. The gold-standard
fix is to put `SelectFromModel` inside a `Pipeline` so it is re-fit on each
training fold independently. For a project of this size the effect is small
(selection is fairly stable across folds for tree-based importance), so this
notebook leaves it as-is but flags it explicitly rather than silently
overstating the CV score. If you want the rigorous version, wrap it like:

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel

pipe = Pipeline([
    ("selector", SelectFromModel(RandomForestClassifier(random_state=42, n_jobs=-1), threshold="mean")),
    ("model", RandomForestClassifier(random_state=42, n_jobs=-1))
])
cross_val_score(pipe, X_train, y_train, cv=3, scoring="roc_auc")
```


````python
# ==========================
# Compare Before and After Feature Selection
# ==========================

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

score_before = cross_val_score(
    rf,
    X_train,
    y_train,
    cv=3,
    scoring="roc_auc",
    n_jobs=-1
).mean()

score_after = cross_val_score(
    rf,
    X_train_selected,
    y_train,
    cv=3,
    scoring="roc_auc",
    n_jobs=-1
).mean()

selection_results = pd.DataFrame({
    "Experiment":[
        "Before Feature Selection",
        "After Feature Selection"
    ],
    "Number of Features":[
        X_train.shape[1],
        X_train_selected.shape[1]
    ],
    "ROC_AUC":[
        score_before,
        score_after
    ]
})

selection_results
````

**feature selection note**
 Feature selection reduced the number of features from 426 to 165. Although this simplified the model, it resulted in a slight decrease in ROC-AUC. Therefore, the final model was trained using all extracted feature

````python
plt.figure(figsize=(6,5))

plt.bar(
    selection_results["Experiment"],
    selection_results["ROC_AUC"]
)

plt.ylabel("ROC-AUC")

plt.title("Effect of Feature Selection")

plt.grid(alpha=0.3)

plt.ylim(.8, .85)
# FIX: filename had no extension and contained spaces
plt.savefig("figures/effect_of_feature_selection.png", dpi=300, bbox_inches="tight")

plt.show()
````

## Hyperparameter Tuning & Final Model


````python
# ==========================================================
# Hyperparameter Tuning
# ==========================================================

from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(
    random_state=42,
    n_jobs=-1
)

param_grid = {
    "n_estimators":[
        200,
        300
    ],
    "max_depth":[
        20,
        None
    ],
    "min_samples_split":[
        2,
        5
    ],
    "min_samples_leaf":[
        1,
        2
    ],
    "max_features":[
        "sqrt"
    ]
}
````

````python
grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    scoring="roc_auc",
    n_jobs=-1,
    verbose=2
)

# NOTE: trained on the full 426-feature set by design (see the feature
# selection note above) - selection cost ~0.4pp ROC-AUC for a smaller model,
# and the full feature set was chosen here to prioritize raw performance.
grid.fit(
    X_train,
    y_train
)
print("Best Parameters")

print(grid.best_params_)

print()

print("Best ROC-AUC")

print(grid.best_score_)
````

````python
# Final model

best_rf = grid.best_estimator_

y_pred = best_rf.predict(
    X_test
)

y_prob = best_rf.predict_proba(
    X_test
)[:,1]
#================================
# Classification report
#================================
from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        y_pred,
        target_names=[
            "Non-Enzyme",
            "Enzyme"
        ]
    )
)
#==================
# ROC_AUC
#==================
from sklearn.metrics import roc_curve, roc_auc_score
auc = roc_auc_score(
    y_test,
    y_prob
)

print(
    "Final ROC-AUC:",
    auc
)
#===============
#CONFUSION MATRIX
#================
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(
    y_test,
    y_pred
)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=[
        "Non-Enzyme",
        "Enzyme"
    ]
)

disp.plot(
    cmap="Blues"
)

plt.title(
    "Confusion Matrix"
)

plt.savefig("figures/confusion_matrix.png", dpi=300, bbox_inches="tight")

plt.show()
````

````python
# ROC curve

fpr, tpr, _ = roc_curve(
    y_test,
    y_prob
)

plt.figure(figsize=(7,6))

plt.plot(
    fpr,
    tpr,
    linewidth=2,
    label=f"Random Forest (AUC={auc:.3f})"
)

plt.plot(
    [0,1],
    [0,1],
    "--",
    color="gray"
)

plt.xlabel(
    "False Positive Rate"
)

plt.ylabel(
    "True Positive Rate"
)

plt.title(
    "ROC Curve"
)

plt.legend()

plt.grid(alpha=.3)

# FIX: filename had no extension and contained spaces
plt.savefig("figures/roc_curve.png", dpi=300, bbox_inches="tight")

plt.show()
````

````python
# Feature importance
#-------------------


importance = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": best_rf.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

top_n = importance.head(20)

plt.figure(figsize=(10,7))

plt.barh(
    top_n["Feature"],
    top_n["Importance"]
)

plt.gca().invert_yaxis()

plt.xlabel(
    "Importance"
)

plt.title(
    "Top 20 Important Features"
)

plt.savefig("figures/feature_importance.png", dpi=300, bbox_inches="tight")

plt.show()
````

#  Error Analysis

This section analyzes the misclassified samples produced by the final model.

````python
# ==========================================================
# Error Analysis
# ==========================================================

    
errors = X_test.copy()

errors["True_Label"] = y_test.values
errors["Predicted_Label"] = y_pred

misclassified = errors[
    errors["True_Label"] != errors["Predicted_Label"]
]

print("Number of misclassified samples:", len(misclassified))
print("Percentage:",
      round(len(misclassified)/len(errors)*100,2), "%")

misclassified.head()
````

## Bonus: Trying a Stronger Model (XGBoost) — Same Features, No Rework

Everything above is untouched. This section is added purely as an extra
experiment: same `X_train` / `X_test` / `y_train` / `y_test` (the full
426-feature set, per the decision above) already in memory, just a
different, generally stronger model on top of them.

Random Forest got **77% accuracy / 0.845 ROC-AUC**. Gradient-boosted trees
(XGBoost/LightGBM) usually pick up an extra 1-3 points on tabular
classification problems like this one because they correct errors
sequentially instead of averaging independent trees. It won't rival adding
PSSM or protein-language-model embeddings (that's the real ceiling-breaker —
see the discussion above), but it's the fastest win available without
touching the feature set.

Requires `xgboost` (`pip install xgboost`) — the cell below checks for it and
tells you clearly if it's missing instead of failing silently.

````python
# ==========================================================
# Bonus: XGBoost on the same selected features
# ==========================================================

try:
    from xgboost import XGBClassifier
    HAS_XGB = True
except ImportError:
    HAS_XGB = False
    print("xgboost is not installed. Run:  pip install xgboost")

if HAS_XGB:

    xgb_model = XGBClassifier(
        n_estimators=300,
        max_depth=6,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        eval_metric="logloss",
        random_state=42,
        n_jobs=-1
    )

    xgb_model.fit(
        X_train,
        y_train
    )

    y_pred_xgb = xgb_model.predict(
        X_test
    )

    y_prob_xgb = xgb_model.predict_proba(
        X_test
    )[:, 1]

    print(
        classification_report(
            y_test,
            y_pred_xgb,
            target_names=["Non-Enzyme", "Enzyme"]
        )
    )

    auc_xgb = roc_auc_score(
        y_test,
        y_prob_xgb
    )

    print("XGBoost ROC-AUC:", auc_xgb)
````

````python
# ==========================
# Compare Random Forest vs XGBoost
# ==========================

if HAS_XGB:

    from sklearn.metrics import accuracy_score

    model_comparison = pd.DataFrame({
        "Model": ["Random Forest (tuned)", "XGBoost"],
        "Accuracy": [
            accuracy_score(y_test, y_pred),
            accuracy_score(y_test, y_pred_xgb)
        ],
        "ROC_AUC": [
            auc,
            auc_xgb
        ]
    })

    print(model_comparison)

    fig, axes = plt.subplots(1, 2, figsize=(11, 5))

    axes[0].bar(
        model_comparison["Model"],
        model_comparison["Accuracy"]
    )
    axes[0].set_title("Accuracy")
    axes[0].set_ylim(0.70, 0.90)
    axes[0].tick_params(axis="x", rotation=15)
    axes[0].grid(alpha=0.3)

    axes[1].bar(
        model_comparison["Model"],
        model_comparison["ROC_AUC"]
    )
    axes[1].set_title("ROC-AUC")
    axes[1].set_ylim(0.75, 0.90)
    axes[1].tick_params(axis="x", rotation=15)
    axes[1].grid(alpha=0.3)

    plt.tight_layout()

    plt.savefig(
        "figures/rf_vs_xgboost.png",
        dpi=300,
        bbox_inches="tight"
    )

    plt.show()

    # ROC curve overlay for a direct visual comparison
    fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_prob_xgb)

    plt.figure(figsize=(7, 6))
    plt.plot(fpr, tpr, linewidth=2, label=f"Random Forest (AUC={auc:.3f})")
    plt.plot(fpr_xgb, tpr_xgb, linewidth=2, label=f"XGBoost (AUC={auc_xgb:.3f})")
    plt.plot([0, 1], [0, 1], "--", color="gray")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve: Random Forest vs XGBoost")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.savefig("figures/roc_curve_comparison.png", dpi=300, bbox_inches="tight")
    plt.show()
````

## Honest Takeaway on the RF vs. XGBoost Comparison

The result above is **XGBoost underperforming the tuned Random Forest**
(Accuracy 0.746 vs 0.769, ROC-AUC 0.829 vs 0.845) — not the improvement one
might expect from a gradient-boosted model. This is a real, reproducible
result from this run, not an error, and it's worth stating plainly rather
than glossing over:

- **The comparison isn't fair yet.** `best_rf` went through a 48-fit
  `GridSearchCV` search; the `XGBClassifier` above used one fixed,
  reasonable-guess set of hyperparameters with no search at all. An untuned
  model losing to a tuned one is expected, not evidence that XGBoost is worse
  as an algorithm for this problem.
- **Takeaway for anyone reading this notebook:** algorithm choice matters
  less than most people assume until every candidate gets a comparable amount
  of tuning effort. The fair next step — not done here — would be to give
  XGBoost its own `GridSearchCV`/`RandomizedSearchCV` pass over
  `n_estimators`, `max_depth`, `learning_rate`, and `subsample` before
  declaring a winner.
- The tuned Random Forest (ROC-AUC 0.845, Accuracy 0.77) remains the best
  model produced in this notebook and is the one used for the feature
  importance and SHAP analysis above.
